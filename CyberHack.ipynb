Import file and libraries

```python
from google.colab import files
import pandas as pd
import io

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
  # Read the Excel file into a pandas DataFrame
  try:
    log_df = pd.read_excel(io.BytesIO(uploaded[fn]))
    print("\nSuccessfully loaded the data.")
    display(log_df.head())
  except Exception as e:
    print(f"\nAn error occurred while reading the Excel file: {e}")
```

Count total

```python
# Get the total number of rows in the DataFrame
total_rows = len(log_df)

# Count non-null values and calculate null percentages
column_info = pd.DataFrame({
    'Column Name': log_df.columns,
    'Non-Null Count': log_df.count().values,
    'Null Count': log_df.isnull().sum().values,
    'Total Count': total_rows,
    'Null Percentage': (log_df.isnull().sum().values / total_rows) * 100
})

display(column_info)
```

Print all the unique values in each column with count for each

```python
for col in log_df.columns:
    print(f"Value counts for column: {col}")
    display(log_df[col].value_counts())
    print("-" * 30)
```

Breakdown the time column - "TimeCreated"

```python
# Convert 'TimeCreated' to datetime objects if it's not already
log_df['TimeCreated'] = pd.to_datetime(log_df['TimeCreated'])

# Extract date, hour, minute, and second from 'TimeCreated'
log_df['Date'] = log_df['TimeCreated'].dt.date
log_df['Hour'] = log_df['TimeCreated'].dt.hour
log_df['Minute'] = log_df['TimeCreated'].dt.minute
log_df['Second'] = log_df['TimeCreated'].dt.second

print("Timestamp components added to the DataFrame.")
display(log_df[['TimeCreated', 'Date', 'Hour', 'Minute', 'Second']].head())
```

Plot for hours-minute for 10:15, 10:16 and 10:17

```python
import matplotlib.pyplot as plt

# Create a new column combining hour and minute
log_df['Hour_Minute'] = log_df['TimeCreated'].dt.strftime('%H:%M')

# Group by hour and minute and count events
events_per_hour_minute = log_df.groupby('Hour_Minute').size()

# Plot the data
plt.figure(figsize=(15, 7))
events_per_hour_minute.plot(kind='line')
plt.title('Number of Events per Hour and Minute')
plt.xlabel('Time (Hour:Minute)')
plt.ylabel('Number of Events')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
